```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

LX, LY = 1.2, 1.2

gridwidth = 0.2
X, Y = np.meshgrid(np.arange(-LX, LX, gridwidth), np.arange(-LY, LY, gridwidth))

R = np.sqrt(X**2 + Y**2)

# Coordonnées de position et charge de deux charges ponctuelles
X1, Y1 = 0, 0.03
Q1 = 0.8
R1 = np.sqrt((X - X1) ** 2 + (Y - Y1) ** 2)
ax.plot(X1, Y1, "o", color="blue")

X2, Y2 = 0, -0.03
Q2 = -Q1
R2 = np.sqrt((X - X2) ** 2 + (Y - Y2) ** 2)
ax.plot(X2, Y2, "o", color="blue")
##

# Paramètres de la fonction vectorielle. 2 charges.
U = Q1 * (X - X1) / (R1**1.3) + Q2 * (X - X2) / (R2**1.3)
V = Q1 * (Y - Y1) / (R1**1.3) + Q2 * (Y - Y2) / (R2**1.3)

M = np.sqrt(U**2 + V**2)        # norme de la croissance 

ax.quiver(X, Y, U/M, V/M, M, pivot='mid');
```
