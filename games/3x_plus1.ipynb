# 3n + 1

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def syracuse_seq(start):
    current = start
    sequence = []

    while current >1:
        sequence.append(current)
        if not current % 2: # even case
            current = current // 2
            
        else: # odd case
            current = 3*current +1
    return sequence
```

```python
len(syracuse_seq(3413))
```

```python
start = 12083

fig, ax = plt.subplots(figsize=(20,10))

lengths = []
for n in range(start, start+4, 1):
    seq = syracuse_seq(n)
    lengths.append(len(seq))
    ax.scatter(x=range(len(seq)), y=seq, label = f"Start: {n}, max: {max(seq)}, len: {len(seq)}")
    ax.text(x=seq.index(max(seq)) - .1 , y=max(seq) +2, s=f'{max(seq)}')
    ax.plot(seq)
    
ax.legend()
ax.set_xlabel('Length of the sequence')
ax.set_ylabel('Number reached')
ax.set_xticks(np.arange(0, max(lengths) +1, max(lengths) // 7 ));
```

```python
seq1 = syracuse_seq(580)
seq2 = syracuse_seq(581)
arr1 = np.array(seq1)
arr2 = np.array(seq2)
```
